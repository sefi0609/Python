```python
from time import sleep
#Introduction to the game + print index board
def introduction():
    
    print('Hello and Welcom to the Tic Tac Toe game!!!')
    print("In this game first you need to choose who will be X and who will be O")
    print("Then you need to select who will start - 'X' or 'O'")
    print("Every step of the way you will need to choose a number that indicate")
    print("The place you chose to put you X or O")
    print("The board look's like this: ")
    print_index_board()
    print("Okay so now let's start to play :-)")
    sleep(5)
```

```python
from IPython.display import clear_output
```

```python
#Select who is the first player
def first_player():
    
    global XorO
    
    while True:
        
        player = input("Please select who the first player is  - 'X' or 'O': ").upper()
        
        if player == 'X':
            XorO = True
            break
            
        elif player == 'O':
            XorO = False
            break
            
        else:
            print('Sorry, invalid input, Please try again')    
```

```python
#Input an index of the board from the user, can't repeat an index
def input_index():
    
    global indexes_list
    
    while True:
        
        #sleep(2)
        #clear_output()
        
        #print the new board
        #print_board(board)
        
        index = input('Please enter a number for the place you have choose (1-9): ')
        
        if index not in indexes_list:
            
            if index.isdigit():
                
                if int(index) in range(1,10):
                    
                    indexes_list.append(index)
                    return int(index)
                
                else:
                    print("Sorry, need a number between 1 to 9")
            else:
                print("Sorry, need a number")
        else:
            print("Sorry, invalid move, Please try again")
```

```python
#Print the index board
def print_index_board():
    
    print('      |     |     ')
    print(f'   {7}  |  {8}  |  {9} ')
    print('      |     |     ')
    print('  ---------------')
    print('      |     |     ')
    print(f'   {4}  |  {5}  |  {6} ')
    print('      |     |     ')
    print('  ---------------')
    print('      |     |     ')
    print(f'   {1}  |  {2}  |  {3} ')
    print('      |     |     ')    
```

```python
#print a new board every Iteration 
def print_board(board):

    clear_output()
    print('      |     |     ')
    print(f'   {board[6]}  |  {board[7]}  |  {board[8]} ')
    print('      |     |     ')
    print('  ---------------')
    print('      |     |     ')
    print(f'   {board[3]}  |  {board[4]}  |  {board[5]} ')
    print('      |     |     ')
    print('  ---------------')
    print('      |     |     ')
    print(f'   {board[0]}  |  {board[1]}  |  {board[2]} ')
    print('      |     |     ')
```

```python
#Check if there is a winner
def check_board():
    
    global game_on
    global indexes_list
    
    if len(set(board[:3])) == 1:
        
        if 'X' in board[:3]:
            print("Game Over, Player 'X' won thanks for playing")
            game_on = False
            
        elif 'O' in board[:3]:
            print("Game Over, Player 'O' won thanks for playing")
            game_on = False
            
        else:
            pass
        
    if len(set(board[3:6])) == 1:
        
        if 'X' in board[3:6]:
            print("Game Over, Player 'X' won thanks for playing")
            game_on = False
            
        elif 'O' in board[3:6]:
            print("Game Over, Player 'O' won thanks for playing")
            game_on = False
            
        else:
            pass
        
    if len(set(board[6:9])) == 1:
        
        if 'X' in board[6:9]:
            print("Game Over, Player 'X' won thanks for playing")
            game_on = False
            
        elif 'O' in board[6:9]:
            print("Game Over, Player 'O' won thanks for playing")
            game_on = False
            
        else:
            pass

    #Check all diagonals
    if len(set(board[::4])) == 1:
        
        if 'X' in board[::4]:
            print("Game Over, Player 'X' won thanks for playing")
            game_on = False
            
        elif 'O' in board[::4]:
            print("Game Over, Player 'O' won thanks for playing")
            game_on = False
    
        else:
            pass
        
    if len(set(board[2:7:2])) == 1:
        
        if 'X' in board[2:7:2]:
            print("Game Over, Player 'X' won thanks for playing")
            game_on = False
            
        elif 'O' in board[2:7:2]:
            print("Game Over, Player 'O' won thanks for playing")
            game_on = False
            
        else:
            pass
    
    #Check all columns
    if len(set(board[:7:3])) == 1:

        if 'X' in board[:7:3]:
            print("Game Over, Player 'X' won thanks for playing")
            game_on = False
            
        elif 'O' in board[:7:3]:
            print("Game Over, Player 'O' won thanks for playing")
            game_on = False
            
        else:
            pass
    
    if len(set(board[1:8:3])) == 1:
        
        if 'X' in board[1:8:3]:
            print("Game Over, Player 'X' won thanks for playing")
            game_on = False
            
        elif 'O' in board[1:8:3]:
            print("Game Over, Player 'O' won thanks for playing")
            game_on = False
            
        else:
            pass
        
    if len(set(board[2::3])) == 1:
        
        if 'X' in board[2::3]:
            print("Game Over, Player 'X' won thanks for playing")
            game_on = False
            
        elif 'O' in board[2::3]:
            print("Game Over, Player 'O' won thanks for playing")
            game_on = False
            
        else:
            pass

    #If the board is full
    if len(indexes_list) == 9:
        
        print("Game Over, no onegame_on won :-(")
        game_on = False      
```

```python
#insert an object - 'O' or 'X' to the board
def insert_object(index):
    
    global XorO
    
    if XorO == True: #Need to insert 'X'
        
        XorO = False
        board[index-1] = 'X'
    
    else: #Need to insert 'O'
        XorO = True
        board[index-1] = 'O'
```

```python
#check if the users want to play again
def replay():
    
    choose = input('Do you want to play again? enter Y/N: ')
    
    return choose == 'Y'
```

```python
#reset the board for the new game
def reset_board():
    
    global board
    global indexes_list
    
    indexes_list = []
    board = [' ']*9
```

```python
#main routine
game_on = True #if game is on than True else False
XorO = True #if 'X' than True else False
indexes_list = [] #this is a list of occupied indexes

#introduction to the game
introduction()
#choose how is the first player
first_player()
#reset the board and the indexes_list
reset_board()

while game_on:
    #input an index 
    index = input_index()
    #insert an object
    insert_object(index)
    #Print a new board every Iteration
    print_board(board)
    #check if there's a winner
    check_board()
    #check if the players want to keep playing
    if game_on == False:
        replay()
```
